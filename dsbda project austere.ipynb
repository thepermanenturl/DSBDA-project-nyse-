```python
def commonintwo(a,b):
    c=[]
    for i in a:
        if i in b:
            c.append(i)
    return c
```

```python
def commoninn(argv):
    n=len(argv)
    if n==1:
        return argv[0]
    elif n==2:
        return commonintwo(argv[0],argv[1])
    else:
        return commonintwo(commoninn(argv[:n-1]),argv[-1])
```
